```python
# Scenario:
# You are organizing an online webinar. 
# Many people register, but some may register multiple times by mistake.
#  You want to keep a list of unique attendees only.
```

```python
n = int (input ("Enter the number of participants"))
participants = set()

for i in range (n):
    name = input ("Enter the names of participants")
    participants.add(name)

print("all",participants)



# for i in range(len(participants)):
# #    for j in range(i+1, len(participants)):
#     for j in range (len(participants)-1, i ,-1):
#        if participants[i] == participants[j]:
#               participants.pop(j)



              
           
```
