```python
# Scenario:
# You are a teacher keeping track of students’ exam scores. Each student has a unique name, and each student’s score needs to be recorded. Later, you want to:

# Check a student’s score

# See all students with their scores
```

```python
#used dictionary to store unique names
student_names = {"Alice", "Bob"}
student_profile = {}

for i in student_names:
    score = (input("Enter the score for " + i ))
    student_profile[i] = score

print(student_profile)

search_name = input("Enter the name to check their score: ")

f = student_profile[search_name]
print("Score of " + search_name + " is " + f)

        
#  Questions- How to lowercase name

    

   


    
```
