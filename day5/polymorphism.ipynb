```python
# #polymorphism - many forms
# same function naame behaves differenty for different objects

# 1 function different behaviour
```

```python
# eg:car, bike truck, 
# common function -> start
# car start w key, bike w button

# eg: notification 
# does send by  email, sms
```

```python
class vehicle:
    def start(self):
        print("vehicle is starting")

class Car(vehicle):
    def start(self):
        print("Car start w key")

class Bike(vehicle):
    def start(self):
        print("start w button")

class Truck(vehicle):
    def start(self):
        print("start w engine")

#polymorphism in action
vehicle = [vehicle(),Car(), Bike(), Truck()]

for v in vehicle:
    v.start()
```
