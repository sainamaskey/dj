```python
# inheritance
# one class can use property and method of another class
# #child can use property of parent class
```

```python
# eg: school system where
# parent class -> person
# child class -> student
# ko class of person(name, age) - student can take
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_info(self):
        print (f"Name: {self.name}, Age: {self.age}")
    
class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age) # calling parent constructur
        self.roll_no = roll_no

    def show_student(self):
        print (f'Roll {self.roll_no}')

# Creating objects
s1 = Student("Raj", 33,43)
s1.show_info() #inherited method
s1.show_student() # own method
```

```python
# #ecommerce
# parent -> product
# child -> electronics, clothing
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def show_info(self):
        print (f"Name: {self.name}, Price: {self.price}")
    
class Electronics(Product):
    def __init__(self, name, price, warenty):
        super().__init__(name, price ) # calling parent constructur
        self.warenty = warenty

    def show_detail(self):
        print (f'warenty {self.warenty}')

class Clothing(Product):
    def __init__(self, name, price, size):
        super().__init__(name, price ) # calling parent constructur
        self.size = size

    def show_detail(self):
        print (f'size {self.size}')

# Creating object
laptop = Electronics("Laptop",423,5)
laptop.show_info() #inherited method
laptop.show_detail() # own method
# s1.show_sizes() # own method

shirt = Clothing("shirt",423,5)
shirt.show_info() #inherited method
shirt.show_detail() # own method
```
