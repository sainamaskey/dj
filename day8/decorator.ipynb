```python
def My_decorator(func):
    def wrapper():
        print("before calling")
        func()
        print("ater call")
    return wrapper

@My_decorator
def say_hello():
    print("hello")

say_hello()
```

```python
def require_login(func):
    def wrapper(user):
        if user!="admin":
            print("access denied")
        else: 
            func(user)
    return wrapper

@require_login
def view_dashboard(user):
    print (f"{user} viewing")

view_dashboard("raju")
view_dashboard("admin")
```
