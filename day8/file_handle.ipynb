```python
# to keep record of product, quantity, total price

j =0
sm=0
info = ""
n = int (input("enter no product"))

while j<n:
    p = input("enter product")
    c =int( input ("enter quantity"))
    f = int(input ("enter price"))
    sm = sm+c*f
    info = info + f"{p},{c},{f}, {sm}\n"
    j = j+1
file = open("bill.csv",'w' )
file.write("name,price, quantity, total \n")
file.write(info)
file.close()
```
