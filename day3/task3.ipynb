```python
# ATM Withdrawal
# Scenario:ATM allows withdrawal only if balance is sufficient.
# Question:Create a function that checks if withdrawal is possible.
```

```python
def withdrawl ( amount, balance):
    if amount <= balance:
        return "withdrawl success"
    else:
        return "not enough balance"
    
amount = float(input("Enter amount: "))
balance = float(input("Enter balance: "))

withdrawl(amount=amount, balance=balance)
```
