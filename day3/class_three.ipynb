```python
# No param, No return
def great():
    print ("hello welcom")

great()
```

```python
#  param, No return
def great(name):
    print ("hello welcom", name)

great("Ram")
```

```python
# no param,  return
def great():
  return 10

great()
```

```python
#  param,  return
def add(a, b):
  return a+b

add(2,3)
```

```python
# lamda function (short function)- 1 line function

add = lambda a,b:a+b
print(add(2,3))

```

```python
#variable length arguments
# can take multiple
# to take multiple args

def add(*numbers):
    print (sum(numbers))
add(1,2,3,5)
```

```python
def total(*args):
    print(args)

total(1,2,3,4,4)
```

```python
def sum_all(*numbers):
    total = 0
    for n in numbers:
        total +=n 
    print(total)

sum_all(12,23,34,34)
```
