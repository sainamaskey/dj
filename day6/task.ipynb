```python
# Design a School Management System using Object-Oriented Programming concepts.

# Create a base class Person
# Create a derived class Student
# Demonstrate the use of:
# Public members
# Protected members
# Private members
# Requirements:
# Store student name as public
# Store school name as protected
# Store student password as private
# Provide a method to verify login using password
# Show how protected data is accessed in child class
# Show that private data cannot be accessed directly
```

```python
class Person:
    def __init__(self, student_name, school_name, student_password):
         self.student_name = student_name  
         self._school_name = school_name  
         self.__student_password = student_password  

    def verifypassword(self, username, password):
        if username.lower() == self.student_name.lower() and password == self.__student_password:
               print ( "User Verified")
        else:
             print ("Verification Failed")


class Student(Person):
      
      def showValues ( self):
        # print(f"{self.student_name} from {self._school_name}")
        print(self.student_name, "from", self._school_name)
        # print(self.student_name + " from " + self._school_name)

      def accessuser(self, username, password):
        self.verifypassword(username, password)

s1 = Student("Ram", "Saraswati School", "Ram123")
s1.showValues()
s1.accessuser("raj", "notraj")
s1.accessuser("Ram", "Ram123")
    
```
