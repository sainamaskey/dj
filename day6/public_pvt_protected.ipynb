```python
# public- accessible from everywhere
# syntax used : name

# protected - accessible inside class + child class
#syntax : _name

#private- accessible only inside same class
#syntax: used __name
#makes new variable, doent modify original
```

```python
#example- design library system w inheritance

#base class, derived class
# public- title,
# protected- author, 
# private- price

# method: public- display info, 
# protected- apply discount, 
# pvt- final price
```

```python
class LibraryItems: #parent class
    def __init__(self, title, author, price):
        self.title = title  
        self._author = author  
        self.__price = price  

    def display(self):
        print("title", self.title)
        print("author",self._author)
        print("price", self.__calculate_price())

    def _apply_discount(self):
        return self.__price*0.9
    
    def __calculate_price(self):
        return self._apply_discount()
    
#child class
class Book(LibraryItems):
    def __init__(self, title, author, price, pages):
        super().__init__(title, author, price)
        self.pages = pages

    def show_details(self):
        print("\nBook Details")

        #access public
        print("Title", self.title)
        print("Pags", self.pages)  

        #access protected
        print("Author", self._author) 

        #access private cant do   
        # print(self.__price)  

        self.display()

#Object
b1 = Book("Amazing World", "Ram", 322, 2333)
b1.show_details()

b1 = LibraryItems("AmazWorld", "Ram", 2)
b1.display()
```

```python
class Account: #parent class
    def __init__(self, name, balance, pin):
        self.name = name    # Public member
        self._balance = balance    # Protected member
        self.__pin = pin   # Private member

    def deposit(self, amount):
        self._balance += amount

    def withdraw (self, amount, pin):
        if pin == self.__pin:
            if amount < self._balance:
                self._balance -= amount
                print('withdraw successful')
            else:
                print("not enough balance")
        else: 
            print("wrong pin")
    
    def show_balance(self):
        print("balance", self._balance)
    
#child class
class SavingAccount(Account):
   
    def add_interest(self):
      self._balance += self._balance * 0.05

#Object
b1 = SavingAccount("Ram", 322, 1212)
b1.add_interest()
# print(b1.name)
print(b1._balance) #bad banners, dont follow encapsulation

# b1 = Account("Raj", 434, 1212)
b1.deposit(12)
b1.show_balance()
b1.withdraw(2, 1212)

b1.__pin = 999
print(b1.__pin) #actual pin is still unchanged
b1.withdraw(2, 1212)
```
