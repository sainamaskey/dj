Design a class for a Bank Account system using OOP concepts. The class should:
<!-- Store account holder name and balance -->
Allow deposit and withdrawal
Prevent withdrawing more than balance
Create objects for two customers and perform operations

```python
class Bank: 
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    def withdrawl( self, amount):
        if amount > self.balance: 
            print("cant, not enough")
        else:
            self.balance = self.balance - amount
            print(self.balance, self.name)

    def deposit( self, amount):
        self.balance +=  amount
        print(self.balance, self.name)

    
    # print(balance)
    

action1 = Bank("Raj",5)
action2 = Bank("Kuza",100)

# action1.withdrawl()
action1.withdrawl(4)
action1.deposit(2)
action2.withdrawl(4)
action2.deposit(2)
```
