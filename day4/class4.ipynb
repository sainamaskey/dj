```python
def student_info(**kwargs):
    print (kwargs)

student_info(name = "ran", age = 3)
```

```python
def display(**data):
    for key, value in data.items():
        print(key, value)
display(name = "site", age = 3)
```

```python
class Car:
    def __init__(self,brand,color): #constructor
        self.brand=brand
        self.color=color

    def show(self): #methods is function inside a class
        print(f"{self.brand} car is {self.color}")

car1=Car("Tesla", "Red")

car1.show()
```
