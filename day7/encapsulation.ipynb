```python
from abc import ABC, abstractmethod

#Abstract class
class ElectricityBill(ABC):
    def __init__(self, customer_name, units):
        self.customer_name = customer_name 
        self._units = units
        self.__unit_rate = 5
        self.__tax = 0.10

    @abstractmethod
    def calculate_bill(self):
        pass

    #private method fro internal calc
    def __base_amount(self):
        return self._units * self.__unit_rate
    
    #protected method
    def _apply_tax(self):
        return self.__base_amount() *  (1+self.__tax)
    
#child class 1
class CommercialBill(ElectricityBill):
    def calculate_bill(self):
        return self._apply_tax()
    
#objects -> variable w data, function, method all
# variable -> only 1 type data
com_bill = CommercialBill("ABC ltd", 500)

print ( "Commercial bil", com_bill.calculate_bill())

    
```
