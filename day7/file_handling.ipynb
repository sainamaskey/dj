```python
#File handling
# syntax

# file=open(<file_name>,<mode>)
# file.close()

# mode->
# read -> r
# create -> x
# write -> w -does create and write
# append -> a
```

```python
try: 
    file=open('data.txt', 'x')
    file.close()
except:
    print("cant")
```

```python

file=open('data.txt', 'r')
data = file.read()
print(data)
file.close()
```

```python
with open('data.txt', 'r')as file:
    print (file.read()) # same as above
```
