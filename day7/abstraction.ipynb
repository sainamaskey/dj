```python
# Design a hospital system where treatment abstract,
# child classes surgery and therapy
```

```python
from abc import ABC,abstractmethod

#abstract class
class Treatment(ABC):

    @abstractmethod #decorator 
    def perform_treatment(self):
        pass

#child class 1
class Surgery(Treatment):
    def perform_treatment(self):
        print ("performing w doctors")

#child class 2
class Therapy(Treatment):
    def perform_treatment(self):
        print("perform therpy")

t1 = Surgery()
t2 = Therapy()

t1.perform_treatment()
t2.perform_treatment()

```
