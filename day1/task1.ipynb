```python

item = []
for i in range (5):
    items = input ("enter the shopping list item " + str(i+1))
    item.append(items)

print ("shoppinglist is ", item)
required_item = input("What item do you want to check? ")
if required_item in item:
    print ("item ", required_item, " is present")
else:
    print ("item "+required_item+" is not present")
```
