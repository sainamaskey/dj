```python
item = []
price_list = []
quantity_list = []
total_cost = 0
item_price = []

for i in range (2):
    items = input ("enter the shopping list item " +str(i+1))
    item.append(items)


print ("shoppinglist is ", item)

for i in item:
        prices  = float(input ("enter price of "  + i )) 
        price_list.append(prices)
        quantity = int(input ("enter quantity of " + i))
        quantity_list.append(quantity)
        item_price.append(prices * quantity)

#for i in item:
for i in range(len(item)):
      print ("item ->", item[i], "price ", price_list[i], "-> quantity is ", quantity_list[i])
      total_cost += price_list[i] * quantity_list[i]

print ("\n")
for i in range(len(item)):
      print ("the total cost of ", item[i], " is ", item_price[i])

print ("\ntotal bill amount is ", total_cost)

search_item = input ("enter the item you want to search ")
if search_item in item:
    print ("item ", search_item, " is present")
else:    print ("item "+search_item+" is not present")
```
